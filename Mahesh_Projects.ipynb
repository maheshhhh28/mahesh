# **Product Category Classifier for ShopEase E-Commerce**

```python
# Import library
import pandas as pd
```

```python
# Read dataset
df = pd.read_csv('/content/Product Category Classifier for ShopEase E-Commerce.csv')
```

```python
df.head()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.utils import to_categorical

# Drop rows with missing target or key features
df = df.dropna(subset=['description', 'brand', 'color', 'size', 'material', 'price', 'target_category'])

# Encode target labels
le = LabelEncoder()
df['category_encoded'] = le.fit_transform(df['target_category'])
y = to_categorical(df['category_encoded'])

# Preprocessing for features
text_features = 'description'
categorical_features = ['brand', 'color', 'size', 'material']
numeric_features = ['price']

# Define column transformer
preprocessor = ColumnTransformer(transformers=[
    ('text', TfidfVectorizer(max_features=100), text_features),
    ('cat', OneHotEncoder(handle_unknown='ignore'), categorical_features),
    ('num', StandardScaler(), numeric_features)
])

# Apply preprocessing
X_processed = preprocessor.fit_transform(df)

# Train/test split
X_train, X_test, y_train, y_test = train_test_split(X_processed, y, test_size=0.2, random_state=42)

# Input dimension for ANN
input_dim = X_train.shape[1]
output_dim = y.shape[1]

# Define ANN model
model = Sequential()
model.add(Dense(128, activation='relu', input_dim=input_dim))
model.add(Dropout(0.3))
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.2))
model.add(Dense(output_dim, activation='softmax'))

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])

# Train the model
history = model.fit(X_train.toarray(), y_train, epochs=10, batch_size=32, validation_split=0.1, verbose=0)
```

```python
# Evaluate model
loss, accuracy = model.evaluate(X_test.toarray(), y_test, verbose=0)
accuracy
```
